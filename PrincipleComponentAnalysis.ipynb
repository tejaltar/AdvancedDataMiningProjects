```python
import pandas as pd
# download dataset file and keep it in same folder named as iris.data
data = pd.read_csv('iris.data', names=['sepal_length','sepal_width','petal_length','petal_width','class'])
data
```

```python
attributes = ['sepal_length','sepal_width','petal_length','petal_width']
xAxis = data.loc[:, attributes].values
xAxis = pd.DataFrame(xAxis)
xAxis
```

```python
yAxis = data.loc[:,['class']].values
yAxis[yAxis == 'Iris-setosa'] = 'm'
yAxis[yAxis == 'Iris-versicolor'] = 'c'
yAxis[yAxis == 'Iris-virginica'] = 'k'
yAxis = pd.DataFrame(yAxis, columns = ['class'])
yAxis
```

```python
from sklearn.decomposition import PCA
pcaComponents = PCA(n_components=2)
components = pcaComponents.fit_transform(xAxis)
dataFrame = pd.DataFrame(data = components, columns = ['Column 1', 'Column 2'])
dataFrame
```

```python
newData = pd.concat([dataFrame, yAxis], axis = 1)
newData
```

```python
import matplotlib.pyplot as matplot
newData.plot.scatter('Column 1', 'Column 2', c = newData['class'])
```

```python
import numpy as np
import pandas as pd
import random
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics.pairwise import manhattan_distances
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

xs = list()
ys = list()
euclideanList = list()
manhattanList = list()

for i in range(1,500):
    x = random.randint(100,1000)
    y = random.randint(1,100)
    xs.append(x)
    ys.append(y)
    data = np.random.randint(1,100,size = (x,y))
    data = pd.DataFrame(data)
    euclidean = euclidean_distances(data, data)
    euclidean[euclidean == 0] = np.nan
    euclidean = pd.DataFrame(euclidean)
    euclideanMax = euclidean.max(axis = 1)
    euclideanMax = euclideanMax.max(axis = 0)
    euclideanMin = euclidean.min(axis = 1)
    euclideanMin = euclideanMin.min(axis = 0)
    euclideanPlot = (euclideanMax - euclideanMin) / euclideanMin
    euclideanList.append(euclideanPlot)
    manhattan = manhattan_distances(data, data)
    manhattan[manhattan == 0] = np.nan
    manhattan = pd.DataFrame(manhattan)
    manhattanMax = manhattan.max(axis = 1)
    manhattanMax = manhattanMax.max(axis = 0)
    manhattanMin = manhattan.min(axis = 1)
    manhattanMin = manhattanMin.min(axis = 0)
    manhattanPlot = (manhattanMax - manhattanMin) / manhattanMin
    manhattanList.append(manhattanPlot)

fig = plt.figure()
euclidean = fig.add_subplot(111, projection='3d')

euclidean.scatter(xs, ys, euclideanList, c='c', marker='o')

euclidean.set_xlabel('N')
euclidean.set_ylabel('D')
euclidean.set_zlabel('Gamma')

plt.show()
```

```python
fig = plt.figure()
manhattan = fig.add_subplot(111, projection='3d')
manhattan.scatter(xs, ys, manhattanList, c='b', marker='o')

manhattan.set_xlabel('N')
manhattan.set_ylabel('D')
manhattan.set_zlabel('Gamma')

plt.show()
```

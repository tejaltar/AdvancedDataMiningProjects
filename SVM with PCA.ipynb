```python
import pandas as pd
import numpy as np
import bz2
from bz2 import decompress
from liblinearutil import *
from datetime import datetime
fileData = bz2.open("C:/Users/tejal/Downloads/Assignment 2/covtype.libsvm.binary.bz2", 'rb')
file = fileData.read()
```

```python
y, x = svm_read_problem("C:/Users/tejal/Downloads/Assignment 2/covtype.libsvm.binary")
```

```python
import array
a = array.array('i',(i for i in range(0,54)))
df = pd.DataFrame(data = x, columns = a)
df = df.fillna(0)
num = df.iloc[:,1:11]
meanVal = num.mean(axis=0)
maxVal = num.max(axis=0)
for i in range(1,11):
    df[i] = (df[i] - meanVal[i]) / maxVal[i]
del df[0]
df.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.5, random_state=42, stratify=y)
```

```python
from sklearn.decomposition import PCA
pcaComponents = PCA(n_components=10)
components = pcaComponents.fit_transform(X_train)
dataFrame = pd.DataFrame(components)
dataFrame
```

```python
pcaComponents = PCA(n_components=10)
components = pcaComponents.fit_transform(X_test)
dataFrame_test = pd.DataFrame(components)
dataFrame_test
```

```python
x_train = dataFrame.to_dict('records')
x_test = dataFrame_test.to_dict('records')
```

```python
now = datetime.now()
current_time = now.strftime("%H:%M:%S")
print(current_time)
```

```python
prob = problem(y_train, x_train)
param = parameter('-s 2')
model = train(prob, param)
p_label, p_acc, p_val = predict(y_test, x_test, model)
(Acc, MSE, SCC) = evaluations(y_test, p_label)
print(Acc, MSE, SCC)
```
